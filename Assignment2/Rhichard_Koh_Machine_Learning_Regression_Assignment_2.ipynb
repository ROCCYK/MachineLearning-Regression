#Normal Equation

```python
import numpy as np
def normal_eq_lm(X,y):
  X_np = X.to_numpy()
  y_np = y.to_numpy()
  ones = np.ones(len(X))
  X_np = np.concatenate((ones[:, np.newaxis], X_np), axis=1)
  Xt = X_np.transpose()
  first_part = np.dot(Xt,X_np)
  first_part = np.linalg.inv(first_part)
  second_part = np.dot(Xt, y_np)
  return np.dot(first_part,second_part)

def normal_eq_lm_single(X,y):
  X_np = X.to_numpy()
  y_np = y.to_numpy()
  ones = np.ones(len(X))
  X_np = np.reshape(X_np,(len(X),1))
  X_np = np.concatenate((ones[:, np.newaxis], X_np), axis=1)
  Xt = X_np.transpose()
  first_part = np.dot(Xt,X_np)
  first_part = np.linalg.inv(first_part)
  second_part = np.dot(Xt, y_np)
  return np.dot(first_part,second_part)
```

#Gradient Descent

```python
def gd_lm(X, y, Θ, b, alpha=0.01):
  # Θ = 0.0
  # b = 0.0
  derivative_Θ = 0.0
  derivative_b = 0.0
  N = X.shape[0]
  for xi,yi in zip(X,y):
    derivative_Θ += -2*xi*(yi-(Θ*xi+b))
    derivative_b += -2*(yi-(Θ*xi+b))
  Θ = Θ - alpha * (1/N) * derivative_Θ
  b = b - alpha * (1/N) * derivative_b
  return Θ, b
```

##For Horse Power

```python
w = 0.0
b = 0.0
for epoch in range(1000):
  Θ, b = gd_lm(X['hp'],y,w,b)
  y_hat = Θ * X + b
  MSE = np.divide(np.sum((y - y_hat)**2, axis=0), X.shape[0])
  print(f'{epoch} weight: {Θ} bias: {b}')
```

```python
def loss(y,hat):
    N = len(y)
    sum = 0
    for i in range(N):
        sum += (y[i]-hat[i])**2
    return (1/N)*sum
```

```python
def y_hat(w,x,b):
    pred_list = []
    for i in range(len(x)):
        pred_list.append(np.dot(w,x[i])+b)
    return np.array(pred_list)
```

```python
def dldw(x,y,y_hat):
    sum = 0
    N = len(y)
    for i in range(N):
        sum += -x[i]*(y[i]-y_hat[i])
    return (2/N)*sum
```

```python
def dldb(y,y_hat):
    N = len(y)
    sum = 0
    for i in range(N):
        sum += -(y[i]-y_hat[i])
    return (2/N) * sum
```

```python
def gd_lm(x,y, epoch=400, alpha=0.01):
    w = np.random.randn(x.shape[1])
    b = 0
    for i in range(epoch):
        y_predicted = y_hat(w,x,b)
        w = w - alpha *dldw(x,y,y_predicted) 
        b = b - alpha * dldb(y,y_predicted)    
    return w,b
```
